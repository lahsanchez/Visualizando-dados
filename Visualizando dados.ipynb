```
import numpy as np
from matplotlib import pyplot as plt
```

### Carregando DataSet ####

```
data = np.genfromtxt('/content/drive/MyDrive/dataset/iris.data.txt', delimiter=',', usecols=(0,1,2,3))
```

```
print(data[0:,0])
```

```
data[50:100]
```

### Comparando iris Sertosa com iris Versicolor ###
### Montando grafico de comparação do comprimento da Sépala ###

```
irisSertosa = data[0:50,0]
irisVersicolor = data[50:100,0]
xs = [i for i, _ in enumerate(irisSertosa)]
plt.plot(xs, irisSertosa, 'rs:', label='iris-sertosa')
plt.plot(xs, irisVersicolor, 'b.:', label='iris-versicolor')

plt.legend(loc=1)
plt.title('Comparando Iris do comprimento da Sepala')
plt.xlabel('Dados linhas')
```

### Comparando iris Sertosa com iris Virginica ###
### Montando grafico de comparação do comprimento da Sépala ###

```
irisSertosa = data[0:50,0]
irisVirginica = data[100:150,0]
xs = [i for i, _ in enumerate(irisSertosa)]
plt.plot(xs, irisSertosa, 'rs:', label='iris-sertosa')
plt.plot(xs, irisVirginica, 'c.:', label='iris-virginica')

plt.legend(loc=1)
plt.title('Comparando Iris do comprimento da Sepala')
plt.xlabel('Dados das linhas')
```
